```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Data: [Voltage, Current]
X = np.array([
    [5,0.5],[5.5,0.4],[4.8,0.6],[6,0.3],[6.5,0.2],
    [6.2,0.25],[4.9,0.55],[5.8,0.35],[5.2,0.5],[6.3,0.22]
])

# Labels: Working=0, Defective=1
y = np.array([0,0,0,1,1,1,0,1,0,1])

# Model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

# Test sample
test = np.array([[5.7, 0.38]])
pred = model.predict(test)
print("component Status:","Working" if pred[0]==0 else "Defective")

import matplotlib.pyplot as plt
plt.scatter(X[y==0][:,0], X[y==0][:,1], label="Working")
plt.scatter(X[y==1][:,0], X[y==1][:,1], label="Defective")
plt.scatter(test[:,0], test[:,1], marker='x', s=100, label="Test")
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.legend()
plt.grid()
plt.show()
```
