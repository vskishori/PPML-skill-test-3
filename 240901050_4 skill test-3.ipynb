```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]).reshape(-1,1)
y = np.array([25,28,30,34,37,40,43,45,47,50])

# Model
model = LinearRegression()
model.fit(X, y)

# Predict price for a future year
year = np.array([[2022]])
pred = model.predict(year)
print("Predicted Price (₹ Lakhs):", pred[0])

import matplotlib.pyplot as plt
plt.scatter(X, y, label="Actual Data")
plt.plot(X, model.predict(X), label="Regression Line")
plt.xlabel("Year")
plt.ylabel("Price (₹ Lakhs)")
plt.legend()
plt.grid()
plt.show()
```
