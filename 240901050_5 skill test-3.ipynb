```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# Encode Transmitter: T1=0, T2=1, T3=2
# Encode Output: Good=1, Bad=0

# Data: [Transmitter, Signal Strength, Frequency]
X = np.array([
    [0,75,2000],
    [1,50,1500],
    [0,90,2200],
    [2,40,1300],
    [1,85,2100],
    [0,70,1800],
    [2,30,1200],
    [1,60,1600],
    [0,80,1900],
    [2,20,1100]
])

# Labels
y = np.array([1,0,1,0,1,1,0,0,1,0])

# Model
model = DecisionTreeClassifier()
model.fit(X, y)

# Test sample: T2, strength=78, freq=2000
test = np.array([[1,78,2000]])
pred = model.predict(test)
print("Good" if pred[0]==1 else "Bad")

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(8,6))
plot_tree(model, feature_names=["Transmitter","Strength","Frequency"],
          class_names=["Bad","Good"], filled=True)
plt.show()
```
