```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Data: [Voltage, Current]
X = np.array([
    [4.819941,0.87293],
    [5.249786,0.918358],
    [4.771151,0.958775],
    [4.602712,1.087788],
    [4.814919,0.877739],
    [12.38281,2.257141],
    [12.36406,2.383657],
    [11.22958,1.89107],
    [12.17975,1.82507],
    [12.0638,2.29575]
])

# KMeans model (2 clusters)
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Print cluster labels
print("Cluster Labels:", labels)
print("Centroids:\n", centers)


plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], marker='x', s=200)
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("K-Means Clustering (Voltage vs Current)")
plt.grid()
plt.show()
```
